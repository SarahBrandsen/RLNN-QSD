```python
import finalfig as ff
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib as tikz
```

```python
#n=50 

total_trials = 3
training_trials = 1500
d = 2*np.ones(50)
m = 3
q = np.array([1/3, 1/3, 1/3])
train_check = 10

bigvec = ff.fulltest(total_trials, training_trials, d, m, q, train_check, evaluation_trials = 5000, SDP = False)
```

```python
import tikzplotlib as tikz
Xvec = train_check*np.arange(int(training_trials/train_check)+1)
for _ in range(3):
    plt.plot(Xvec, bigvec[0][_], linewidth = 1)
plt.xlabel('training iteration')
plt.ylabel('rew_{RLNN}')
tikz.save('Fig13c')
plt.show()
```
