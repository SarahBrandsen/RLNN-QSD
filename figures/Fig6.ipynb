```python
import finalfig as ff
import numpy as np

total_trials = 5
train_check = 125
d = np.array([2,2,2, 2])
m = 3
q = np.array([1/3, 1/3, 1/3])
training_trials = 125

bigvec, vec_SDP, vec_local_SDP, _ = ff.fulltest(total_trials, training_trials,  d, m, q, train_check, SDP = True, LG = False, local_SDP = True, dep = True)
```

```python
import matplotlib.pyplot as plt
import tikzplotlib as tikz


vec_NN = bigvec[:,-1]
vec_SDP, vec_NN, vec_local_SDP = np.array([vec_SDP]).flatten(), np.array([vec_NN]).flatten(), np.array([vec_local_SDP]).flatten()

arrinds = vec_SDP.argsort()
sorted_SDP = vec_SDP[arrinds]
sorted_NN = vec_NN[arrinds]
sorted_localSDP = vec_local_SDP[arrinds]

X = np.arange(total_trials)
plt.xlabel('trial')
plt.ylabel('Success Probability')
plt.plot(X, sorted_SDP, marker = "s", color = 'green', linewidth = 0, label = 'SDP')
plt.plot(X, sorted_NN, color='black', marker = "o", linewidth=0, label = 'RLNN')
plt.plot(X, sorted_localSDP, marker = "x", color = 'blue', linewidth = 0, label = 'local SDP')
plt.legend(loc="lower right")
tikz.save("Fig5")
plt.show()
```
